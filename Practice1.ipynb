```python
# import necessary packages 
import pandas as pd
from urllib.parse import quote_plus
from sqlalchemy import create_engine
from sqlalchemy import text

password = input("Enter the Password")
```

```python
# establish connection with the database 
engine = create_engine(f"""mysql+mysqlconnector://root:{quote_plus(password)}@localhost:3306/world""")
```

```python
# read table data using sql query 
sql_df=pd.read_sql_query(text("SELECT * FROM city"), con=engine.connect())
sql_df
```

```python
# Access Diffrent Database SQL Query 
pd.read_sql_query(text("""select country, avg(creditLimit) as avg_credit 
from classicmodels.customers 
group by country 
having avg_credit > 80000;"""), con=engine.connect())
```

```python
# csv file access from pandas
data = pd.read_csv("pokemon_data.csv")
data
```

```python
# Data Insersion from csv file to MYSQL Database(Auto Commit)
data.to_sql("pokemon_data",engine.connect(),index=False,if_exists="replace")
```

```python
# save the csv file in another csv file
data.to_csv("test1_csv",index=False)
```
